```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## An introduction

**Trends** - a trend is defined as a pattern of change. **Line charts** are best to show trends over a period of time, and multiple lines can be used to show trends in more than one group.

**Relationship:**

- **Bar charts** are useful for comparing quantities corresponding to different groups.
- **Heatmaps** can be used to find color-coded patterns in tables of numbers.
- **Scatter plots** show the relationship between two continuous variables; if color-coded, we can also show the relationship with a third categorical variable.
- **Categorical scatter plots** show the relationship between a continuous variable and a categorical variable.

Including **a regression line** in the scatter plot makes it easier to see any linear relationship between two variables. **sns.lmplot** - This command is useful for drawing multiple regression lines, if the scatter plot contains multiple, color-coded groups.

**Distribution** - we visualize distributions to show the possible values that we can expect to see in a variable, along with how likely they are.

- **Histograms** show the distribution of a single numerical variable.
- **KDE plots (or 2D KDE plots)** show an estimated, smooth distribution of a single numerical variable (or two numerical variables).

**sns.jointplot** - This command is useful for simultaneously displaying a 2D KDE plot with the corresponding KDE plots for each individual variable.

## Sets

**The style of the figure** - sns.set_style("dark")

## Relationships

```python
path = 'https://drive.google.com/uc?export=download&id=1avjiuZB-z87t8HrKiHBOWC3-2OmReaQw'
candy = pd.read_csv(path, index_col='id')
candy.head()
```

### sns.barplot

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=candy, x='chocolate', y='pricepercent')
plt.title('How to differ pricepercent of candies with chocolate and without?')
```

### sns.heatmap

```python
# Data for this plot is pivot table where avarage winpercent is calculated for each pair sugarpercent-pricepercent
plt.figure(figsize=(10,5))
sns.heatmap(data=pd.pivot_table(data=candy, values='winpercent', index='pricepercent', columns='sugarpercent', aggfunc='mean'))
plt.title('Price / sugar amount preferences')
```

### sns.scatterplot

```python
sns.scatterplot(data=candy, x='sugarpercent', y='winpercent')
plt.title("The relationship between 'sugarpercent' and 'winpercent'")
```

### sns.regplot

```python
plt.figure(figsize=(14, 6))
plt.title('Relationship between sugar percent and users scores')
sns.regplot(x=candy['sugarpercent'], y=candy['winpercent'])
```

### sns.lmplot

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=candy, x='pricepercent', y='winpercent', hue='chocolate')
plt.title("The relationship between 'pricepercent' and 'winpercent' saparally for candies with chocolate and without")
```

```python
plt.figure(figsize=(10, 6))
sns.lmplot(data=candy, x='pricepercent', y='winpercent', hue='chocolate')
```

### sns.swarmplot

```python
plt.figure(figsize=(10, 5))
plt.title('Scores of candys with chocolate and without')
sns.swarmplot(data=candy, x='chocolate', y='winpercent')
```

## Distributions

### histplot

```python
sns.histplot(data=candy, x=candy['pricepercent'], hue=candy['peanutyalmondy'])
plt.title("Distribution in values for 'pricepercent' separatly for candies with nuts and without" )
```

### kdeplot

```python
plt.figure(figsize=(10, 6))
plt.title('Distribution of pricepercent')
sns.kdeplot(data=candy, 
            x='pricepercent', 
            hue='peanutyalmondy', 
            shade=True)
```
